```python
qcvcpool = traindice['Pool_QC'].value_counts()
```

```python
def pool_convert(x):
    if x == 'Gd':
        return 3
    elif x == 'Fa':
        return 1
    elif x == 'TA':
        return 2
    elif x == 'Ex':
        return 4
    elif x not in qcvcpool:
        return 0
```

```python
traindice['Pool_QC'] = traindice['Pool_QC'].map(pool_convert)
```

```python
traindice['Pool_QC'].value_counts()
```
